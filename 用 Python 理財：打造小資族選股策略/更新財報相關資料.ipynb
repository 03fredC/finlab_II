# 每季財報 時間區間補充：

關於每季財報爬取，選取的時間範圍要「包含」以下時段，就能夠爬到截止日期的財報囉！例如：

* 2010年3月31號 --> 2009 第四季
* 2010年5月15號 --> 2010 第一季
* 2010年8月14號 --> 2010 第二季
* 2010年11月14號 --> 2010 第三季

所以假如我們想要抓2018年的第一季，就可以輸入一個時間範圍，例如：

* 2018年5月01號 到 2018年5月31號

因為這段時間有包含 2018年5月15號，所以就會下載 2018年第一季的財報喔！
